```python
 #Sorting the top 10 cities based on the SmartCity_Index
top_cities = path.sort_values(by = 'SmartCity_Index', ascending = False).head(10)
top_cities
```

Displaying the heatmap for these TOP 10 Cities 

```python

# Select the columns for the heatmap
heatmap_data = top_cities[['City', 'Smart_Mobility ', 'Smart_Environment', 'Smart_Government ', 'Smart_Economy ', 'Smart_People', 'Smart_Living']]

# Set the city names as the index
heatmap_data.set_index('City', inplace=True)

# Create the heatmap
plt.figure(figsize=(15, 8))
sns.heatmap(heatmap_data, cmap='coolwarm', annot=True, fmt=".1f")

# Show the plot
plt.title('Heatmap of Smartness in Different Categories for Top 10 Cities')
plt.xticks(rotation=0)
plt.show()

```

```python
df =  path[['City', 'SmartCity_Index']].sort_values(by='SmartCity_Index', ascending=False).head(10)
    #Set the city column as index
df.set_index('City', inplace=True)

# Create the heatmap
plt.figure(figsize=(15, 6))
sns.heatmap(df, annot=True, fmt=".1f", cmap="coolwarm")
plt.title('Smartness of Top 10 Cities with SmartCity_Index ')
plt.xlabel('Smartness Metrics')
plt.ylabel('City')
plt.xticks(rotation=0)
plt.show()
```
